Getting the data

```
import numpy as np
import pandas as pd
import math
from google.colab import files

url = 'https://raw.githubusercontent.com/badriadhikari/AI-2020fall/master/activities/bm25.csv'
dataframe= pd.read_csv(url, delimiter=',')

dataframe.head()
```

Set our k and b

```
k = 1.2
b = 0.75
```

change dataframe into numpy darray

```
data = dataframe.values
```

IDF

```
def IDF(q):
    return math.log((data.shape[0] - DF(q) + 0.5) / (DF(q) + 0.5))
```

DF

```
def DF(q):
    total = 0
    for i in range(data.shape[0]):
        if data[i, q] > 0:
            total += 1
    return total
```

L

```
def L():
    return np.average(data[:, 1])
```

TF

```
def TF(documentIndex, q):
    return data[documentIndex, q]
```

Algo for BM25

```
def BM25(documentID):
    total = 0
    for i in range(2):
        q = i + 2
        numerator = IDF(q) * ((TF(documentID, q) * (k+1)))
        denominator = TF(documentID, q) + k * (1 - b + b * (np.absolute(data[documentID, 1]) / L()))
        total += (numerator/denominator)
    return total
```

Create array of BM25

```
BM25array = []
for i in range(data.shape[0]):
    BM25array.append(BM25(i))
```

Create numpy array of BM25 array

```
BM25arrnp = np.array(BM25array)
BM25Dataset = np.column_stack((data, BM25arrnp))
```

Sort the Scores

```
sortBM25 = BM25Dataset[BM25Dataset[:, 4].argsort()]

first_and_last_column = sortBM25[:, [0, 4]]
print(first_and_last_column)
```
