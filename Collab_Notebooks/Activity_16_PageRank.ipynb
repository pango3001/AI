````
class Page:
    def __init__(self, score, in_link, out_link):
        self.score = score
        self.in_link = in_link  
        self.out_link = out_link 

def PageRank(main_page):
    total = 0
    for mpage in main_page:
        for inlink in mpage.in_link:
            for inpage in inlink:
                total += inpage.score / inpage.out_link
        mpage.score = ( ((1 - d)/N) + (d * total) )


N = 4
d = 0.7

page1, page2, page3, page4 = [[] for _ in range(N)]

prA = 1.0 # all page ranks are 1.0 at the beginning

#
page1.append( Page(prA, [page3], 2) )
page2.append( Page(prA, [page1, page3], 1) )
page3.append( Page(prA, [page1, page4], 3) )
page4.append( Page(prA, [page3, page2], 1) )

# Updates the page ranks of all the pages
for i in range(128):
    PageRank(page1)
    PageRank(page2)
    PageRank(page3)
    PageRank(page4)

# Output
print("Page1 score: ", page1[0].score)
print("Page2 score: ", page2[0].score)
print("Page3 score: ", page3[0].score)
print("Page4 score: ", page4[0].score)
````
